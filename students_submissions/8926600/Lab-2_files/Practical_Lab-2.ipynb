```python
import matplotlib.pyplot as plt
countries = ['China', 'India', 'USA']
gdp = [14.34, 2.87, 22.68]
plt.bar(countries, gdp, color=['blue', 'green', 'red'])
plt.title('Comparison of GDP (2023)')
plt.xlabel('Country')
plt.ylabel('GDP (Trillions of USD)')
plt.show()
```

```python
import matplotlib.pyplot as plt
companies = ['Apollo', 'MRF', 'CEAT']
sales = [800, 1200, 600]
plt.pie(sales, labels=companies, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Sales of Three Companies (2023)')
plt.axis('equal') 
plt.show()
```

```python
import matplotlib.pyplot as plt
platforms = ['iOS', 'Android']
users = [1000, 2500]
plt.bar(platforms, users, color=['blue', 'green'])
plt.title('Users of iOS vs. Android (2023)')
plt.xlabel('Platform')
plt.ylabel('Number of Users (Millions)')
plt.show()
```

```python
import matplotlib.pyplot as plt
years = [2018, 2019, 2020, 2021, 2022, 2023]
growth_rate = [5, 7, 10, 8, 12, 9]
plt.plot(years, growth_rate, marker='o', linestyle='-', color='purple')
plt.title('Cosmetics Sector Growth (2018-2023)')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.grid(True)
plt.show()
```
